```python
# set dataframe ke bentuk yang bisa diproses apriori
basket = df.groupby(['Transaction', 'Item'])['Item'].count().unstack().fillna(0)
basket = basket.applymap(lambda x: 1 if x > 0 else 0)
basket.head()
```

```python
# mencari itemset yang sering muncul
frequent_items = apriori(basket, min_support=0.02, use_colnames=True)
frequent_items.head(100)
```

```python
frequent_items.sort_values(by='support', ascending=False).head(100)
```

```python
multi_itemsets = frequent_items[frequent_items['itemsets'].apply(lambda x: len(x) > 1)]

multi_itemsets.sort_values(by='support', ascending=False).head(100)
```

```python
rules = association_rules(frequent_items, metric="lift", min_threshold=1)
rules = rules.sort_values(by="lift", ascending=False)
rules.head()
```

```python
rules[['antecedents','consequents','support','confidence','lift']].head(10)
```

```python
import matplotlib.pyplot as plt

top_items = frequent_items[frequent_items['itemsets'].apply(lambda x: len(x)==1)]
top_items = top_items.sort_values(by='support', ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top_items['itemsets'].astype(str), top_items['support'])
plt.xticks(rotation=45)
plt.title("Top 10 Most Frequent Items")
plt.ylabel("Support")
plt.show()
```

```python
pair_items = frequent_items[frequent_items['itemsets'].apply(lambda x: len(x)==2)]
pair_items = pair_items.sort_values(by='support', ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(pair_items['itemsets'].astype(str), pair_items['support'])
plt.xticks(rotation=45)
plt.title("Top 10 Frequent 2-Itemsets")
plt.ylabel("Support")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(rules['support'], rules['confidence'], alpha=0.7,
            c=rules['lift'], cmap='viridis')
plt.colorbar(label='Lift')
plt.xlabel("Support")
plt.ylabel("Confidence")
plt.title("Support vs Confidence Colored by Lift")
plt.show()
```

```python
import networkx as nx

G = nx.DiGraph()

for _, rule in rules.iterrows():
    G.add_edge(list(rule['antecedents'])[0],
               list(rule['consequents'])[0],
               weight=rule['lift'])

plt.figure(figsize=(10,8))
pos = nx.spring_layout(G, k=0.4)
nx.draw(G, pos, with_labels=True, node_size=3000,
        node_color='skyblue', font_size=10,
        width=[d['weight'] for (_,_,d) in G.edges(data=True)])
plt.title("Association Rules Network Graph")
plt.show()
```
